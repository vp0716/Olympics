```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv("https://raw.githubusercontent.com/frontiertechinstitute/datasets/refs/heads/main/AIHSRP%20Olympics/athlete_events.csv")

data.head()
```

```python
# 1457 Badminton entries
data = data.dropna(subset=['Age', 'Height', 'Weight'])
```

```python
features = data.columns.tolist()
data = data.drop_duplicates(subset=features)
```

```python
data.shape
```

```python
racket_df = data[data["Sport"].isin(["Badminton", "Tennis", "Table Tennis"])]
# shape is 5159 x 15
```

```python
# number of athletes visualization
sport_counts = racket_df['Sport'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(sport_counts.index, sport_counts.values)
plt.xlabel('Sport')
plt.ylabel('Number of Athletes')
plt.title('Number of Athletes in Each Sport')
plt.show()
plt.savefig("sportcounts.png")
```

```python
# age distribution visualization
plt.figure(figsize=(7,4))
plt.hist(racket_df["Age"], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("age_distribution.png")
plt.show()
```

```python
# medal distribution visualization
pd.crosstab(racket_df["Sport"], racket_df["Medal"]).plot(kind="bar", stacked=True)

plt.title("Medal Distribution Across Racket Sports")
plt.ylabel("Athletes")
plt.tight_layout()
plt.savefig("medals_by_sport.png")
plt.show()
```

```python
racket_df["Medal_Won"] = racket_df["Medal"].notna()
```

```python
racket_df.boxplot(column="Age", by="Medal_Won")

plt.title("Age of Medalists vs. Non-Medalists")
plt.xlabel("Won a Medal")
plt.ylabel("Age")
plt.show()
```

```python
sports = ["Badminton", "Tennis", "Table Tennis"]

for sport in sports:
    subset = racket_df[racket_df["Sport"] == sport]
    plt.scatter(subset["Height"], subset["Weight"], label=sport)

plt.title("Height vs. Weight by Sport")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()

plt.show()
```
